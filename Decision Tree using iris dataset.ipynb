#### Post pruning decision trees with cost complexity pruning
The DecisionTreeClassifier provides parameters such as min_samples_leaf and max_depth to prevent a tree from overfiting. Cost complexity pruning provides another option to control the size of a tree. In DecisionTreeClassifier, this pruning technique is parameterized by the cost complexity parameter, ccp_alpha, which balances the complexity of the tree with its fit to the training data. Greater values of ccp_alpha increase the number of nodes pruned.

cost complexity pruning works by trimming away less important parts of a decision tree, reducing its complexity while preserving its accuracy on new, unseen data.

### Working of cost complexity pruning :


Building a Large Tree:
Initially, a decision tree is built using all the available features and training data. This tree may grow quite large and complex, capturing every detail and noise in the data.

Measuring Tree Complexit
: Cost complexity pruning evaluates the complexity of the tree by considering both its size (number of nodes) and its accuracy on the training data. The more complex the tree (more nodes), the higher the cost. 

Iterative Prun
ng: Cost complexity pruning iteratively prunes (removes) nodes from the tree. It starts from the bottom of the tree and removes nodes that have the least impact on the overall accuracy of the tree. These nodes are pruned based on a cost parameter called alpha (α), which penalizes complexity.

Finding Optimal 
lpha: Cost complexity pruning calculates different values of alpha (α) and evaluates the resulting pruned trees using a validation dataset or through cross-validation. It selects the alpha value that gives the simplest tree while still maintaining good accuracy on unseen 
ata.

Final 
runing: Once the optimal alpha value is determined, the decision tree is pruned using this value, resulting in a simpler and more generalized tree structure that is less likely to overfit to the training data.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier()

#This part computes the complexity pruning path for the decision tree model based on the training data (X_train and y_train).
#ccp_alphas contains the effective alphas for pruning the decision tree.
path = model.cost_complexity_pruning_path(X_train, y_train)
print(path)
ccp_alphas = path.ccp_alphas
print(ccp_alphas)

clfs = []
accuracy = []

#In this loop, a decision tree model is trained for each value of ccp_alpha (effective alpha for pruning).
#The DecisionTreeClassifier is instantiated with random_state=0 to ensure reproducibility, and ccp_alpha is set accordingly.
#The model is then fitted to the training data (X_train, y_train), and the trained model is stored in the clfs list.
#Additionally, the accuracy of each model on the test data (X_test, y_test) is calculated using the accuracy_score and stored in the accuracy list.
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)
    pred = clf.predict(X_test)
    accuracy.append(accuracy_score(y_test, pred))

print(clfs)
print(accuracy)
```

#### Benefits of pruning
Simplicity: Pruning can simplify the decision tree, making it easier to interpret and understand. A simpler model is often preferred because it is more interpretable and easier to communicate to stakeholders.

Reduced OverfittinP pruning can help reduce overfitting by removing unnecessary branches in the tree that capture noise or irrelevant patterns in the training data. This can improve the model's ability to generalize to unseen data.

Computational Efficiency: A pruned tree may require less memory and computational resources for both training and inference, making it more efficient, especially for large datasets or resource-constrained environments.

Robustness: Pruning can make the model more robust to variations and changes in the data. By removing overly complex and specific patterns, the pruned tree may generalize better to new or slightly different datasets.

Feature Importance: Pruning can also help identify the most important features in the dataset by highlighting the splits that are retained in the pruned tree. This can provide insights into the most relevant factors driving the decision-making process.

### Pre-pruning

Pre-pruning, also known as early stopping, works by setting constraints on the tree-building process during training. Unlike post-pruning, which involves growing a full tree and then removing nodes, pre-pruning prevents the tree from becoming overly complex in the first place. 

Pre-pruning is computationally efficient because it avoids the need to grow a full tree before pruning, and it can often lead to faster training times.

However, pre-pruning requires careful tuning of hyperparameters to find the right balance between model complexity and predictive performance.


In the context of pre-pruning decision trees, GridSearchCV and RandomizedSearchCV are techniques used for hyperparameter optimization, including finding the optimal hyperparameters for pre-pruning.

#### GridSearchCV
Grid search is a technique that exhaustively searches through a specified subset of hyperparameter combinations. It creates a grid of all possible hyperparameter values and evaluates the model performance using cross-validation for each combination. This is beneficial for pre-pruning because you can specify a grid of hyperparameters related to pre-pruning constraints . By evaluating the model's performance across various combinations of these parameters, you can determine the optimal settings that prevent overfitting and maximize predictive accuracy.

#### RandomizedSearchCV
Randomized search is an alternative to grid search that randomly samples hyperparameter values from specified distributions. It's beneficial when the search space is large and exhaustive search is not feasible. Randomized search explores different combinations of hyperparameters, evaluating the model's performance using cross-validation. While it may not cover every possible combination, it can efficiently explore the hyperparameter space and identify promising regions that lead to good model performance.
